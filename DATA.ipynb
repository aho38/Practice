# Loading Data in Pytorch

This notebook is meant to highlight some of the ways to load data in pytorch.

## Pictures from file

Suppose your project requires you to use images that are downloaded or in a current directory. For this example we will be using the nintendo dataset found at  http://thumbnailpacks.libretro.com/. This zip file contains nintendo screenshots, title cards and box art.

<img src="imgs/directory.png" width="500"/>

Notice the file structure. Here each subfolder is named according to what it contains. This essentially puts the images in its own class. The imagefolder function recognizes this and loads them accordingly. The following uses this method.

```python
# First we import our packages
%matplotlib inline  
import numpy as np
import os
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torch.autograd import Variable
import torchvision
import torchvision.transforms as transforms
import torchvision.datasets as dsets
```

```python
## Custom Helper Function imshow that will help us view things after normalization
def imshow(img):
    img = img / 2 + 0.5 
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1,2,0)))
```

```python
#Here we will load our images. Note: That we are resizing them to 64 x 64 images

transform = transforms.Compose([
    transforms.Resize((64,64)),
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5,0.5,0.5), std=(0.5,0.5,0.5))
])  

batch_size = 16
dset = dsets.ImageFolder("/Users/omardeguchy/Downloads/Nintendo - Nintendo Entertainment System/", transform=transform)
data_loader = torch.utils.data.DataLoader(dset, batch_size=batch_size, shuffle=True)

print('There are {} images'.format(len(dset)))
```

```python
for i, images in enumerate(data_loader):
    images, labels = images
    imshow(torchvision.utils.make_grid(images))
    break
plt.show()
```

```python
# print the labels vector to see the classes
print(labels)
```

## Creating Data

Suppose I wanted to alter an existing data set in order to create a new dataset. Here we will be using the cifar dataset. 

### Visualization

```python
# First I will load the CIFAR dataset from the pytorch inherent datasets
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.5,0.5,0.5), std=(0.5,0.5,0.5))
])

testset = torchvision.datasets.CIFAR10(root='../../Dropbox/DATA/data/', train=False,
                                       download=False, transform = transform)
print('There are {} images in this test set.'.format(len(testset)))
```

```python
# Take one image and look at i
sample_image = testset[400][0]
print(sample_image.shape)
imshow(sample_image)
```

```python
# Suppose I don't want the color version of the image then I can use #####
transform = transforms.Compose(
[transforms.Grayscale(), #######
 transforms.ToTensor()
])

testset = torchvision.datasets.CIFAR10(root='../../Dropbox/DATA/data/', train=False,
                                       download=False, transform = transform)
print('There are {} images in this test set.'.format(len(testset)))

batch_size = 1
test_loader = torch.utils.data.DataLoader(dataset=testset, batch_size=batch_size, shuffle=False)
print('There are {} batches.'.format(len(test_loader)))
```

```python
# Now if I extract an image 
bw_sample = testset[400][0]
print(bw_sample.shape)
bw_sample = bw_sample.numpy()
print(bw_sample.shape)
# because its one channel we can "squeeze out the first dimension"
bw_sample = np.squeeze(bw_sample,axis =0)
print(bw_sample.shape)
plt.imshow(bw_sample, cmap="gray")
```

### Processing The Dataset

```python
# create testing set

# These lists is where we will be storing the images
Transformed_list = list()
Originals_list = list()


for batch_index, batch in enumerate(test_loader):
    image,_ = batch
    # Convert to array
    im_array = image[0].numpy() # Get rid of the batch size [1 1 32 32] -> [1 32 32]
    
    im_array_sq = np.squeeze(im_array, axis=0)#Get rid of the extra dimension [1 32 32]-> [32 32]
     
    im_array_vec =np.reshape(im_array_sq,32*32)# vectorize image [32 32] -> [1024]
    
    noisy_im = im_array_vec + np.random.normal(0,.05,len(im_array_vec))
    
    Originals_list.append(im_array_vec)
    Transformed_list.append(noisy_im)
    
print('Number of Original Images: {}'.format(len(Originals_list))) 
print('Number of Transformed Images: {}'.format(len(Transformed_list)))
```

```python
# Lets check one of the images as a sanity check
original_sample = Originals_list[400]
transformed_sample = Transformed_list[400]
print(original_sample.shape)
print(transformed_sample.shape)
original_sample = np.reshape(original_sample,(32,32))
transformed_sample = np.reshape(transformed_sample,(32,32))
print(original_sample.shape)
print(transformed_sample.shape)
```

```python
# Display the samples 
plt.figure()
plt.imshow(original_sample, cmap='gray') # Sometimes its helpful to have a colorbar
plt.colorbar()
plt.show()

plt.figure()
plt.imshow(transformed_sample, cmap='gray')
plt.colorbar()
plt.show()
```
